```python
import numpy as np
import pandas as pd

df = pd.read_csv('ames.csv')
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
sorted(df.SalePrice)[-10:]
```

```python
plt.hist(df['Year.Built'])
plt.show()
```

```python
X = df[['Lot.Area','Lot.Frontage','Year.Built','Gr.Liv.Area']]
y = df['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
import pandas as pd
neighborhoods = pd.get_dummies(df['Neighborhood'])
neighborhoods
```

```python
df[df['Garage.Area'].isnull()]
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(df[['Garage.Area']])
df['Garage.Area'] = imp.transform(df[['Garage.Area']])
```

```python
X = df[['Lot.Area','Lot.Frontage','Garage.Area','Year.Built','Gr.Liv.Area']]
X = pd.concat([X,neighborhoods],axis=1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```
